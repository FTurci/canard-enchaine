## Francesco analysis

We show hee how we can proceed to explore the data. For example, we get the nearest contacts for every particle centre. To do so, we fix a cutoff radius `rcut` for our search. 

```python
from scipy.spatial import cKDTree

rcut = centres.radius.mean()*1.2  ## RLJ : frac was 1.1 before
centre_tree = cKDTree(centres.coord)
contact_tree = cKDTree(contacts.coord)
centre_contacts=centre_tree.query_ball_tree(contact_tree, rcut)
```

By the same means, we can easily estimate the radial distribution function `g(r)`.

```python
dr =0.25
r = np.arange(dr,80,dr)
nb = centre_tree.count_neighbors(centre_tree,r, cumulative=False)
nb[0]=0 # ignore self-distances
V = np.prod(centres.coord.ptp(axis=0))
N  = centres.coord.shape[0]
g = nb/(4*np.pi*r**2*dr)*V/(N*(N-1))
plt.plot(r,g); plt.xlabel('r [pixels]'); plt.ylabel('g(r)');
```

Similarly, we can compute the coordination within a larger cutoff, for example around the first minimum of `g(r)`.

```python
rcut = 27
neighs=centre_tree.query_ball_tree(centre_tree,rcut)
coordination = [len(n) for n in neighs]
```

```python
plt.hist(coordination, density=True); plt.xlabel('coordination');plt.ylabel('pdf');
```

## contacts

```python
total_contact = [ ]
for c in centre_contacts:
    if len(c)>0:
        total_contact.append(sum(c))
    else:
        total_contact.append(0)
        
num_contact_all = [ ]
for c in centre_contacts:
    num_contact_all.append(len(c))
        
print('numParticle',len(total_contact))
print('numWithContacts',len([ xx for xx in num_contact_all if xx>0]))
print('mostContacts',np.max(num_contact_all))
print('aveContacts',np.mean(num_contact_all))
```

### make a slice

```python
xCen, yCen, zCen = centres.unpack()
xyzCen = [ [xCen[ii],yCen[ii],zCen[ii]] for ii in range(len(xCen)) ] 
print('xyzCen',np.shape(xyzCen))

boxMin = [ np.min(xyzCen[:][ii]) for ii in range(3) ]
boxMax = [ np.max(xyzCen[:][ii]) for ii in range(3) ]
print('box mins',boxMin,'\nmaxs',boxMax)

#print(len(xyz))
#print(xyz[0])

zDepth = 30
zMin = 230
zMax = zMin + zDepth

xyzSlice = [ cc for cc in xyzCen if cc[indZ] > zMin and cc[indZ] < zMax ]
toPlot = [ np.array([ cc[ii] for cc in xyzSlice]) for ii in range(3) ]
num_contact_slice = [ num_contact_all[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
vol_contact_slice = [ total_contact[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
```

### `plt` plots  slices

```python
def plotStuff(axs,xyzSlice,toPlot,num_contact_slice,vol_contact_slice,zMin,zMax,labs=['','']) :

    panelID  = 0 
    ax = axs[panelID]
    ax.title.set_text('all ({u:.1f} < z < {v:.1f}, diameter ~ 20)'.format(u=zMin,v=zMax))
    ax.scatter(toPlot[0],toPlot[1],c='black',alpha=0.8)
    ax.set_aspect(1.0)

    panelID  += 1

    ax = axs[panelID]
    cScale = np.max(num_contact_slice)
    ax.title.set_text(labs[0]+' (blacker is more, max is {n:d})'.format(n=cScale))
    colorsPlt = np.array( [[0.3,1-num_contact_slice[ii]/cScale,1-num_contact_slice[ii]/cScale] for ii,cc in enumerate(xyzSlice) ] )
    ax.scatter(toPlot[0],toPlot[1],c=colorsPlt,alpha=0.8)
    ax.set_aspect(1.0)

    panelID  += 1

    ax = axs[panelID]
    cScale = np.max(vol_contact_slice)
    ax.title.set_text(labs[1]+' (blacker is more, scale is {n:d})'.format(n=cScale))
    colorsPlt = np.array( [[0.3,1-vol_contact_slice[ii]/cScale,1-vol_contact_slice[ii]/cScale] for ii,cc in enumerate(xyzSlice) ] )
    ax.scatter(toPlot[0],toPlot[1],c=colorsPlt,alpha=0.8)
    ax.set_aspect(1.0)

    panelID  += 1
    
def doSlice(zMin,zDepth) :
    zMax = zMin + zDepth

    coMin = 6
    
    ## make slice
    xyzSlice = [ cc for cc in xyzCen if cc[indZ] > zMin and cc[indZ] < zMax ]
    toPlot = [ np.array([ cc[ii] for cc in xyzSlice]) for ii in range(3) ]
    num_contact_slice = [ num_contact_all[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
    vol_contact_slice = [ total_contact[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
    coord_slice = [ np.maximum(0,coordination[ii]-coMin) for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]

    fig,axs = plt.subplots(1,3,figsize=(15,5))

    plotStuff(axs,xyzSlice,toPlot,vol_contact_slice,coord_slice,zMin,zMax,
              labs=['vol contacts','coordination - {n:d}'.format(n=coMin)])

    plt.savefig('slices{z:d}.png'.format(z=zMin))
    plt.show(fig); 
    plt.close()
    
import matplotlib.colorbar as mcb
def scatterSlice(zMin,zDepth) :
    coMin = 6
    
    ## make slice
    xyzSlice = [ cc for cc in xyzCen if cc[indZ] > zMin and cc[indZ] < zMax ]
    toPlot = [ np.array([ cc[ii] for cc in xyzSlice]) for ii in range(3) ]
    num_contact_slice = [ num_contact_all[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
    vol_contact_slice = [ total_contact[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
    coord_slice = [ np.maximum(0,coordination[ii]-coMin) for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]

    fig,axs = plt.subplots(1,1,figsize=(5,5))

    ax = axs
    
    coord_slice_raw = [ coordination[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
    
    binCoord = 1 + np.max(coord_slice_raw) - np.min(coord_slice_raw) 
    binNumC = 1 + np.max(num_contact_slice) - np.min(num_contact_slice) 
    
    hh = ax.hist2d(num_contact_slice,coord_slice_raw,bins=(binNumC,binCoord),cmap=plt.cm.cividis)
    ax.set_xlabel('numC')
    ax.set_ylabel('coord')
    
    
    fig.colorbar(hh[3])  ## hh[3] gets the data for the colobrar


    #plt.colorbar()
    
    plt.show(fig); 
    plt.close()
```

```python
zDepth=30
doSlice(200-zDepth,zDepth)
doSlice(200,zDepth)
doSlice(200+zDepth,zDepth)
```

### 2d histo numContact and coordination

```python
import scipy.stats

zMin = 200
zMax = zMin+zDepth

scatterSlice(zMin,zDepth)

num_contact_slice = [ num_contact_all[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
coord_slice = [ coordination[ii] for ii,cc in enumerate(xyzCen) if cc[indZ] > zMin and cc[indZ] < zMax ]
rCor = scipy.stats.pearsonr(num_contact_slice,coord_slice)
print('pearson correl r =',rCor[0])
```
